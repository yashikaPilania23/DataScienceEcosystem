```python
import os

# Define the base directory
base_dir = '/content/drive/My Drive/stop_sign_classifier'

# Define the paths for extracted directories
extract_dir_stop = os.path.join(base_dir, 'stop')
extract_dir_not_stop = os.path.join(base_dir, 'not_stop')

# Check if these directories exist
print("Extracted stop images directory:", os.path.isdir(extract_dir_stop))
print("Extracted not_stop images directory:", os.path.isdir(extract_dir_not_stop))
```

```python
from PIL import Image
import os

def is_image_file(file_path):
    try:
        with Image.open(file_path) as img:
            img.verify()  # Verify the image file
        return True
    except (IOError, SyntaxError) as e:
        print(f"File {file_path} is not a valid image. Error: {e}")
        return False

def remove_invalid_images(directory):
    if not os.path.exists(directory):
        print(f"Directory {directory} does not exist.")
        return

    for category in ['stop', 'not_stop']:
        category_dir = os.path.join(directory, category)
        if not os.path.exists(category_dir):
            print(f"Category directory {category_dir} does not exist.")
            continue
        for file_name in os.listdir(category_dir):
            file_path = os.path.join(category_dir, file_name)
            if not is_image_file(file_path):
                os.remove(file_path)
                print(f"Removed invalid image: {file_path}")

# Remove invalid images from extracted directories
remove_invalid_images(extract_dir_stop)
remove_invalid_images(extract_dir_not_stop)
```

```python
def verify_images(directory):
    if not os.path.exists(directory):
        print(f"Directory {directory} does not exist.")
        return

    for category in ['stop', 'not_stop']:
        category_dir = os.path.join(directory, category)
        if not os.path.exists(category_dir):
            print(f"Category directory {category_dir} does not exist.")
            continue
        files = os.listdir(category_dir)
        if not files:
            print(f"No images found in directory: {category_dir}")
        else:
            print(f"Directory {category_dir} contains {len(files)} valid images.")

# Verify images in the directories
verify_images(extract_dir_stop)
verify_images(extract_dir_not_stop)
```

```python
import shutil
import random

def split_and_move_files(src_dir, dst_train_dir, dst_test_dir, train_ratio=0.8):
    for category in ['stop', 'not_stop']:
        category_src_dir = os.path.join(src_dir, category)
        if not os.path.exists(category_src_dir):
            print(f"Source directory {category_src_dir} does not exist.")
            continue
        files = os.listdir(category_src_dir)
        if not files:
            print(f"No files found in directory: {category_src_dir}")
            continue

        random.shuffle(files)
        split_idx = int(len(files) * train_ratio)
        train_files = files[:split_idx]
        test_files = files[split_idx:]

        # Move train files
        for file_name in train_files:
            # Removed extra category in the path join
            shutil.move(os.path.join(category_src_dir, file_name), os.path.join(dst_train_dir, file_name))

        # Move test files
        for file_name in test_files:
            # Removed extra category in the path join
            shutil.move(os.path.join(category_src_dir, file_name), os.path.join(dst_test_dir, file_name))

# Split and move files
split_and_move_files(extract_dir_stop, os.path.join(train_dir, 'stop'), os.path.join(test_dir, 'stop'))
split_and_move_files(extract_dir_not_stop, os.path.join(train_dir, 'not_stop'), os.path.join(test_dir, 'not_stop'))

print("Files moved to train and test directories!")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam

# Define data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

# Define the model
model = Sequential([
    Flatten(input_shape=(150, 150, 3)),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)

print("Model training complete!")
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
def preprocess_image(image_path):
    img = image.load_img(image_path, target_size=(150, 150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions for batch input
    img_array /= 255.0  # Rescale image
    return img_array

# Preprocess the uploaded image or the Google Drive image
for img_name in uploaded.keys():
    img_path = f'./{img_name}'  # Path to the uploaded image
    img = preprocess_image(img_path)
```
