```python
# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 2: Define a Function to Generate Synthetic Data
def generate_synthetic_data(n_samples=1000):
    # Based on the three papers, we'll define variables and their assumed distributions

    # Age: Tuberculosis patients age 18 to 65, based on distribution from literature
    age = np.random.normal(40, 12, n_samples).clip(18, 65)

    # Gender: Assuming 50% male, 50% female based on global statistics
    gender = np.random.choice([0, 1], size=n_samples)  # 0 = Male, 1 = Female

    # Duration of treatment in months (min 2 months, max 9 months)
    treatment_duration = np.random.uniform(2, 9, n_samples)

    # Number of missed doses: Based on adherence estimates from papers
    missed_doses = np.random.poisson(2, n_samples)  # Average missed doses

    # Socioeconomic status (SES): 1 = low, 2 = middle, 3 = high
    ses = np.random.choice([1, 2, 3], p=[0.6, 0.3, 0.1], size=n_samples)

    # Comorbidity (Yes=1, No=0): Proportion of patients with comorbidity
    comorbidity = np.random.choice([0, 1], p=[0.7, 0.3], size=n_samples)

    # Smoking history (Yes=1, No=0): Estimated based on patient history in papers
    smoking_history = np.random.choice([0, 1], p=[0.6, 0.4], size=n_samples)

    # Alcohol use (Yes=1, No=0): Based on patient habits
    alcohol_use = np.random.choice([0, 1], p=[0.4, 0.6], size=n_samples)

    # Target Variable: Dropout (Yes=1, No=0)
    # We assume dropout is influenced by missed doses, SES, comorbidity, etc.
    # Using a logistic regression-based relationship to define the probability of dropout
    dropout_prob = (
        0.5 * missed_doses + 0.2 * (ses == 1) + 0.3 * comorbidity +
        0.1 * smoking_history + 0.15 * alcohol_use + np.random.normal(0, 0.5, n_samples)
    )
    dropout = (dropout_prob > 1.5).astype(int)  # Threshold to convert probability to binary

    # Create a DataFrame
    data = pd.DataFrame({
        'Age': age,
        'Gender': gender,
        'Treatment_Duration': treatment_duration,
        'Missed_Doses': missed_doses,
        'SES': ses,
        'Comorbidity': comorbidity,
        'Smoking_History': smoking_history,
        'Alcohol_Use': alcohol_use,
        'Dropout': dropout
    })

    return data

# Step 3: Generate Synthetic Data
data = generate_synthetic_data(n_samples=1000)
print(data.head())

# Step 4: Visualize the Dataset
sns.pairplot(data, hue='Dropout')
plt.show()

# Step 5: Split the Data into Train and Test Sets
X = data.drop('Dropout', axis=1)
y = data['Dropout']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Train a Logistic Regression Model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Step 7: Train a Random Forest Classifier (Optional)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Step 8: Make Predictions on the Test Set
y_pred_logreg = logreg.predict(X_test)
y_pred_rf = rf.predict(X_test)

# Step 9: Evaluate the Models
print("Logistic Regression Performance:")
print(classification_report(y_test, y_pred_logreg))

print("Random Forest Performance:")
print(classification_report(y_test, y_pred_rf))

# Step 10: Confusion Matrix
cm_logreg = confusion_matrix(y_test, y_pred_logreg)
cm_rf = confusion_matrix(y_test, y_pred_rf)

sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()

# Step 11: Model Accuracy
logreg_accuracy = accuracy_score(y_test, y_pred_logreg)
rf_accuracy = accuracy_score(y_test, y_pred_rf)

print(f"Logistic Regression Accuracy: {logreg_accuracy:.2f}")
print(f"Random Forest Accuracy: {rf_accuracy:.2f}")
```

```python
# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier  # XGBoost Import

# Step 2: Define a Function to Generate Synthetic Data
def generate_synthetic_data(n_samples=1000):
    # Based on the three papers, we'll define variables and their assumed distributions

    # Age: Tuberculosis patients age 18 to 65, based on distribution from literature
    age = np.random.normal(40, 12, n_samples).clip(18, 65)

    # Gender: Assuming 50% male, 50% female based on global statistics
    gender = np.random.choice([0, 1], size=n_samples)  # 0 = Male, 1 = Female

    # Duration of treatment in months (min 2 months, max 9 months)
    treatment_duration = np.random.uniform(2, 9, n_samples)

    # Number of missed doses: Based on adherence estimates from papers
    missed_doses = np.random.poisson(2, n_samples)  # Average missed doses

    # Socioeconomic status (SES): 1 = low, 2 = middle, 3 = high
    ses = np.random.choice([1, 2, 3], p=[0.6, 0.3, 0.1], size=n_samples)

    # Comorbidity (Yes=1, No=0): Proportion of patients with comorbidity
    comorbidity = np.random.choice([0, 1], p=[0.7, 0.3], size=n_samples)

    # Smoking history (Yes=1, No=0): Estimated based on patient history in papers
    smoking_history = np.random.choice([0, 1], p=[0.6, 0.4], size=n_samples)

    # Alcohol use (Yes=1, No=0): Based on patient habits
    alcohol_use = np.random.choice([0, 1], p=[0.4, 0.6], size=n_samples)

    # Target Variable: Dropout (Yes=1, No=0)
    # We assume dropout is influenced by missed doses, SES, comorbidity, etc.
    # Using a logistic regression-based relationship to define the probability of dropout
    dropout_prob = (
        0.5 * missed_doses + 0.2 * (ses == 1) + 0.3 * comorbidity +
        0.1 * smoking_history + 0.15 * alcohol_use + np.random.normal(0, 0.5, n_samples)
    )
    dropout = (dropout_prob > 1.5).astype(int)  # Threshold to convert probability to binary

    # Create a DataFrame
    data = pd.DataFrame({
        'Age': age,
        'Gender': gender,
        'Treatment_Duration': treatment_duration,
        'Missed_Doses': missed_doses,
        'SES': ses,
        'Comorbidity': comorbidity,
        'Smoking_History': smoking_history,
        'Alcohol_Use': alcohol_use,
        'Dropout': dropout
    })

    return data

# Step 3: Generate Synthetic Data
data = generate_synthetic_data(n_samples=1000)
print(data.head())

# Step 4: Visualize the Dataset
sns.pairplot(data, hue='Dropout')
plt.show()

# Step 5: Split the Data into Train and Test Sets
X = data.drop('Dropout', axis=1)
y = data['Dropout']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Train a Logistic Regression Model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Step 7: Train a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Step 8: Train an XGBoost Classifier
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)

# Step 9: Make Predictions on the Test Set
y_pred_logreg = logreg.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_xgb = xgb.predict(X_test)

# Step 10: Evaluate the Models
print("Logistic Regression Performance:")
print(classification_report(y_test, y_pred_logreg))

print("Random Forest Performance:")
print(classification_report(y_test, y_pred_rf))

print("XGBoost Performance:")
print(classification_report(y_test, y_pred_xgb))

# Step 11: Confusion Matrix
cm_logreg = confusion_matrix(y_test, y_pred_logreg)
cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

# Confusion Matrix Visualization
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()

sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues')
plt.title("XGBoost Confusion Matrix")
plt.show()

# Step 12: Model Accuracy
logreg_accuracy = accuracy_score(y_test, y_pred_logreg)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)

print(f"Logistic Regression Accuracy: {logreg_accuracy:.2f}")
print(f"Random Forest Accuracy: {rf_accuracy:.2f}")
print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")
```
