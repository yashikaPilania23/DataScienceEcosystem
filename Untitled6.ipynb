```python
import numpy as np
import pandas as pd
import random

def generate_synthetic_data(num_samples):
    data = []
    for _ in range(num_samples):
        age = random.randint(1, 15)  # Age in months
        head_holding = 1 if age >= 2 else 0
        rolling_over = 1 if age >= 4 else 0
        sitting_with_support = 1 if age >= 6 else 0
        sitting_without_support = 1 if age >= 9 else 0
        crawling = 1 if age >= 9 else 0
        walking_with_support = 1 if age >= 12 else 0
        walking_independently = 1 if age >= 15 else 0

        reaches_for_toys = 1 if age >= 4 else 0
        transfers_objects = 1 if age >= 6 else 0
        pincer_grasp = 1 if age >= 9 else 0
        imitates_scribbling = 1 if age >= 12 else 0

        smiles_response = 1 if age >= 2 else 0
        enjoys_interaction = 1 if age >= 4 else 0
        recognizes_familiar_people = 1 if age >= 6 else 0
        stranger_anxiety = 1 if age >= 9 else 0
        shows_preference = 1 if age >= 12 else 0

        coos = 1 if age >= 2 else 0
        babbles = 1 if age >= 4 else 0
        imitates_sounds = 1 if age >= 6 else 0
        understands_no = 1 if age >= 9 else 0
        says_mama_dada = 1 if age >= 12 else 0

        follows_moving_things = 1 if age >= 2 else 0
        looks_at_hands = 1 if age >= 4 else 0
        explores_objects_with_mouth = 1 if age >= 6 else 0
        finds_partially_hidden_objects = 1 if age >= 9 else 0
        knows_ordinary_things = 1 if age >= 12 else 0

        developmental_delay = int(
            not (head_holding and rolling_over and sitting_with_support and sitting_without_support and
                 crawling and walking_with_support and walking_independently and reaches_for_toys and
                 transfers_objects and pincer_grasp and imitates_scribbling and smiles_response and
                 enjoys_interaction and recognizes_familiar_people and stranger_anxiety and shows_preference and
                 coos and babbles and imitates_sounds and understands_no and says_mama_dada and follows_moving_things and
                 looks_at_hands and explores_objects_with_mouth and finds_partially_hidden_objects and knows_ordinary_things)
        )
        data.append([
            age, head_holding, rolling_over, sitting_with_support, sitting_without_support, crawling, walking_with_support,
            walking_independently, reaches_for_toys, transfers_objects, pincer_grasp, imitates_scribbling,
            smiles_response, enjoys_interaction, recognizes_familiar_people, stranger_anxiety, shows_preference,
            coos, babbles, imitates_sounds, understands_no, says_mama_dada, follows_moving_things, looks_at_hands,
            explores_objects_with_mouth, finds_partially_hidden_objects, knows_ordinary_things, developmental_delay
        ])

    columns = [
        'Age', 'Head Holding', 'Rolling Over', 'Sitting with Support', 'Sitting without Support', 'Crawling',
        'Walking with Support', 'Walking Independently', 'Reaches for Toys', 'Transfers Objects',
        'Pincer Grasp', 'Imitates Scribbling', 'Smiles in Response', 'Enjoys Interaction', 'Recognizes Familiar People',
        'Stranger Anxiety', 'Shows Preference', 'Coos', 'Babbles', 'Imitates Sounds', 'Understands No',
        'Says Mama/Dada', 'Follows Moving Things', 'Looks at Hands', 'Explores Objects with Mouth',
        'Finds Partially Hidden Objects', 'Knows Ordinary Things', 'Developmental Delay'
    ]

    df = pd.DataFrame(data, columns=columns)
    return df

# Generate synthetic data
df = generate_synthetic_data(1000)
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Split the data into features and target
X = df.drop('Developmental Delay', axis=1)
y = df['Developmental Delay']

# Split the data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Print evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
